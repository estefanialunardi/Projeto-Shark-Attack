## Lunar Phases and the attacks

### Do the Lunar Phase interfeer on the attacks?
Researches suggest that shark attacks occur more ore less often depending on phases of the moon. The exact cause remains unclear, but there is reason to believe it's due to the influence of the moon on the sea tides.

### Sea tides and the moon phases
Caused by the moon's (majoritively) and the sun's gravitational forces exerted on the earth, the tidal force generates very long waves that affects the sea level at the coast. When this level is higher, we call it a high tide. When it's lower, the coast is experiencing a low tide.

Every lunar day (or each 24 hours and 50 minutes), there are two low tides and two high tides. Depending on the tidal force, the range between low and high tides are bigger or smaller. 

At new moon or full moon, the tide's range is at its maximum (called spring tide). That mean that high tides are higher and low tides are lower.
At quarter phases, the tide's range is smaller (called neap tide). That mean that high tides are lower and low tides are higher.

Considering this, during the darkest and brightest phases, the tide's variation are much bigger than at the medium luminisoty phases.


```python
#I'll create a new column with the corresponding lunar phase, calculated by the date(year+2.2)*11+day+month/30 - 
#By computing the remainder of the division I'll know the lunar phase:
#0 - 7 - new moon | 8 -14 Waxing | 15 - 21 Full moon - 22 - 29 Waning 

fases_da_lua =[]       
#Convert Calendar Date to Julian Date
for row in range(len(when_attacked.index)):
    day =int(when_attacked.iloc[row]['day'])
    month = int(when_attacked.iloc[row]['month'])
    year = int(when_attacked.iloc[row]['year'])
    
    #Formula to calculate the lunar phase
    try:
        h = year // 100          
        yy = year % 100
        g = (yy + 9) % 19 - 9
        gc = g * 11
        while ( gc < -29 ): 
            gc += 30;
        if (gc > 0): 
            gc %= 30;
        c = {17:7,18:1, 19:-4, 20:-8, 21:16, 22:11, 23:6,24:1, 25:-4, 26:-9, 27:15, 28:11, 29:6, 30:0}[h]
        if month < 3:
            mc = 2
        else:
            mc = 0
        phase = (month + mc + day + gc + c + 30 ) % 30
        if phase < 2:
            fases_da_lua.append("New moon")
        elif phase < 7:
            fases_da_lua.append("Waxing crescent")
        elif phase < 9:
            fases_da_lua.append("First quarter")
        elif phase < 14:
            fases_da_lua.append("Waxing gibbous")
        elif phase < 16:
            fases_da_lua.append("Full moon")
        elif phase < 22:
            fases_da_lua.append("Waning gibbous")
        elif phase < 24:
            fases_da_lua.append("Last quarter")
        elif phase < 29:
            fases_da_lua.append("Waning crescent")
        else:
            fases_da_lua.append("New moon")
    except KeyError:
        fases_da_lua.append("0")
             


#for row in range(len(when_attacked.index)):
#    if when_attacked.iloc[row]['day'] != 0:
#        remainder=(when_attacked.iloc[row]['year']+2.2)*11+int(when_attacked.iloc[row]['day'])+int(when_attacked.iloc[row]['month'])
#        if remainder %30 <8:
#            fases_da_lua.append('New Moon')
#        elif remainder %30 <15:
#            fases_da_lua.append('Waxing Moon')
#        elif remainder %30 <22:
#            fases_da_lua.append('Full Moon')
#        else:
#            fases_da_lua.append('Waning Moon')
#    else:
#        fases_da_lua.append('0')
when_attacked['Lunar Phase'] = fases_da_lua 
when_attacked['Lunar Phase'].value_counts()
```
